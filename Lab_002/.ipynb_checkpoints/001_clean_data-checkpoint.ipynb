```python
# import packages for analysis and modeling
import pandas as pd  # data frame operations
#from pandas.tools.plotting import scatter_matrix  # scatter plot matrix
from pandas import Series

import warnings
warnings.filterwarnings("ignore")
```

```python
OG_DATA = pd.read_csv('Zip_Zhvi_SingleFamilyResidence.csv', encoding = "ISO-8859-1")

OG_DATA.head()
```

```python
OG_DATA.describe()
```

```python
def totalNulls(df):
    initNull = df.isnull().sum()
    total = 0
    for i in initNull:
        if i > 0:
            total = total+i
    return "Total Nulls: {}".format(total)

print(totalNulls(OG_DATA))
print(OG_DATA.shape)
```

```python
data_001 = OG_DATA.dropna()
print(totalNulls(data_001))
print(data_001.shape)
print('Percentage of Data: {}'.format(data_001.shape[0]/OG_DATA.shape[0]))
```

```python
data_001.RegionID = data_001.RegionID.apply(str)
#data_001.loc[:,'RegionID'] = data_001.loc[:,'RegionID'].apply(str)

count = 0
to_keep = []
for c in data_001.RegionID:
    if (len(c) == 5):
        to_keep.append(count)
        
    count=count+1

    
data_002 = data_001.iloc[to_keep]
print(totalNulls(data_002))
print(data_002.shape)

data_002.RegionID = data_002.RegionID.apply(str)
count=0
for c in data_002.RegionID:
    if (len(c) != 5):
        count=count+1

data_003 = data_002.iloc[:,1:]
print('Bad zips: {}'.format(count))
print('Percentage of Data Kept: {}'.format(data_003.shape[0]/OG_DATA.shape[0]))
```

```python
data_003.to_csv('clean_house_data.csv', index = False)
```
